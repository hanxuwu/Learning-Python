```python
class LinkedList:
    def __init__(self, L = None, comparison_key = lambda x: x):
        if L is None or not len(L):
            self.head = None
        else:
            self.head = Node(L[0])
            current_last_node = self.head
            for e in L[1: ]:
                current_last_node.next_node = Node(e)
                current_last_node = current_last_node.next_node
        self.comparison_key = comparison_key
                
    def print_list(self, sep = ', '):
        values = []
        node = self.head
        while node:
            values.append(str(node.value))
            node = node.next_node
        print(sep.join(values))
        
    def append(self, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
            return
        current_node = self.head
        while current_node.next_node:
            current_node = current_node.next_node
        current_node.next_node = new_node
        
    def prepend(self, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
        else:
            new_node.next_node = self.head
            self.head = new_node
            
    def delete_first_element(self):
        if not self.head:
            return
        if not self.head.next_node:
            self.head = None
#        first = self.head
        self.head = self.head.next_node
#        first.next_node = None
        

    def delete_last_element(self):
        if not self.head:
            return
        if not self.head.next_node:
            self.head = None
            return
        node = self.head
        while node.next_node.next_node:
            node = node.next_node
        node.next_node = None
        
    def is_sorted(self):
        if not self.head or not self.head.next_node:
            return True
        node = self.head
        while node.next_node:
            if self.comparison_key(node.value) >\
                         self.comparison_key(node.next_node.value):
                return False
            node = node.next_node
        return True
    
    def reverse(self):
        if not self.head or not self.head.next_node:
            return
        tail = self.head
        current_node = self.head.next_node
        tail.next_node = None
        while current_node:
            next_current_node = current_node.next_node
            current_node.next_node = tail
            tail = current_node
            current_node = next_current_node
        self.head = tail
        
class Node:
    def __init__(self, value):
        self.value = value
        self.next_node = None

        
LL = LinkedList([2, 4, 1, 6])
LL.reverse()
LL.print_list()







```
